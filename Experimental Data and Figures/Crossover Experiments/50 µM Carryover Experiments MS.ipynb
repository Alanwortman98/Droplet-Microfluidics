```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/alanwortman/Desktop/Grad School/Stephenson/Droplet Microfluidics/Experimental Data/Crossover Experiments/FS-ONLY-CARRYOVER-combinedCSV.CSV')


# x_values and y_values are now a series
x1 = df['1,3,5-OMePh time']
y1 = df['1,3,5-OMePh signal']
x2 = df['7-Azaindole time']
y2 = df['7-Azaindole signal']

fig, (ax1, ax2) = plt.subplots(figsize = (10,6), nrows = 2, ncols = 1)

ax1.plot(x1, y1, color = '#58508d', label = '1,3,5-Trimethoxybenzene')
ax1.legend(frameon = False)
ax1.set_title('Fused Silica Capillary (100 m)')
ax1.spines[['right', 'top']].set_visible(False)
ax1.set_ylabel('Abundance')
ax1.set_xlim(0, 6.5)

ax2.plot(x2, y2,  color = '#ff6361', label = '7-Azaindole')
ax2.legend(frameon = False)
ax2.spines[['right', 'top']].set_visible(False)
ax2.set_ylabel('Abundance')
ax2.set_xlabel('Time (Min)')
ax2.set_xlim(0, 6.5)

plt.savefig('FS Capillary Carryover.jpeg', dpi = 300)
```
