```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt

data = pd.read_csv('/Users/alanwortman/Desktop/Grad School/Stephenson/Droplet Microfluidics/Experimental Data/Ru(bpy)3_carry_over_Uv-Vis.csv')
x_data = data['time']
y_data = data['absorbance (460 nm)']

fig, ax = plt.subplots(nrows = 1, ncols = 1)

# plot title uses unicode to specify subscripts
ax.set_title('Uv-Vis Droplet Carryover')
ax.plot(x_data, y_data, color = '#002966', label = 'Ru(bpy)3\u2083Cl\u2082')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Absorbance (460nm)')
ax.legend()
ax.set_xlim(312, 420)
ax.set_ylim(0, 0.8)



im = plt.imread('/Users/alanwortman/Desktop/Final Stephenson group logo.png')
# fig, axis = plt.subplots(figsize = (5,5))
newax = fig.add_axes([0.65, 0.58, 0.2, 0.2], anchor = 'SE')
newax.imshow(im)
newax.axis('off')

plt.show()
```
