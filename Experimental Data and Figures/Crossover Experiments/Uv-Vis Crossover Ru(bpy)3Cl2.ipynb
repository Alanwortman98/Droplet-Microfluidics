```python
data = pd.read_csv('/Users/alanwortman/Desktop/Grad School/Stephenson/Droplet Microfluidics/Experimental Data/Ru(bpy)3_carry_over_Uv-Vis.csv')
x = data['time']
y = data['absorbance (460 nm)']

plt.figure(figsize = (10,6))

ax = plt.subplot()
ax.plot(x, y, color = '#58508d', label = '$Ru(bpy)_{3}Cl_{2}$')
ax.set_ylabel('Absorbance (460 nm)')
ax.set_ylim(0, 1)
ax.set_xlabel('Time (s)')
ax.set_xlim(310, 420)
ax.set_title('UV-Vis Droplet Carryover')
plt.legend(loc = 'upper right', frameon = False)
ax.spines[['right', 'top']].set_visible(False)

plt.savefig('Ru(bpy)3 Carryover UV-Vis.jpeg', dpi = 300)


```
