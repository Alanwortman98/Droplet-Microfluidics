```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import find_peaks

df = pd.read_csv('/Users/alanwortman/Desktop/Grad School/Stephenson/Droplet Microfluidics/\
Experimental Data/Reactions/1 Analyte Trifluoromethylation/Boc-Pyrrole Trifluoromethylation/\
Rain-x_union_with_and_without_light_MS.csv')

x_light = df['with light time']
y_light = df['with light counts']
x_no_light = df['no light time']
y_no_light = df['no light counts']

plt.figure(figsize = (10,6))
ax = plt.subplot()
ax.set_xlim(50, 450)
ax.set_ylim(0, 800000)

sns.lineplot(data = df, x = x_light, y = y_light, color = '#003f5c', label = 'Blue LEDs')
sns.lineplot(data = df, x = x_no_light, y = y_no_light, color = '#ff6361', label = 'No Light', alpha = 0.5)
sns.despine(right = True)
plt.legend(frameon = False)

light_peaks = find_peaks(y_light, height = 50000, distance = 50)

for peak in light_peaks[0]:
    ax.text(x_light[peak], y_light[peak] + 20000, x_light[peak], rotation = 90, horizontalalignment = 'center',\
    color = 'k', fontsize = 11.5)
    
    
no_light_peaks = find_peaks(y_no_light, height = 120000)

for peak in no_light_peaks[0]:
    ax.text(x_no_light[peak], y_no_light[peak] + 20000, x_no_light[peak], rotation = 90, horizontalalignment = 'center',\
    color = 'k', fontsize = 11.5)




```
