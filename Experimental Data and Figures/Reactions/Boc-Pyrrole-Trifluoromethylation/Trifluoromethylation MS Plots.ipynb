```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import find_peaks

MS_PFA = pd.read_csv('/Users/alanwortman/Desktop/Grad School/Stephenson/Droplet Microfluidics/\
Experimental Data/Reactions/\
1 Analyte Trifluoromethylation/Boc-Pyrrole Trifluoromethylation/pyrrole_MS_PFA.csv')

m_z = MS_PFA['m/z']
ion_counts = MS_PFA['ion counts']

plt.figure(figsize = (15,6))
ax = plt.subplot()
sns.lineplot(data = MS_PFA, x = m_z, y = ion_counts, color = '#003f5c', label = 'PFA Tubing')
ax.set_xlim(50, 500)
sns.despine(right = True)
plt.legend(frameon = False)

peaks = find_peaks(ion_counts, height = 5000, distance = 50)

for peak in peaks[0]:
    ax.text(m_z[peak], ion_counts[peak] + 1000, m_z[peak], rotation = 90, horizontalalignment = 'center',\
    color = 'k', fontsize = 11.5)
```
