```python
import matplotlib.pyplot as plt
import pandas as pd

df_MRM = pd.read_csv('/Users/alanwortman/Desktop/Grad School/Stephenson/\
Droplet Microfluidics/Experimental Data/Crossover Experiments/Smiles_Scan_and_4-Cl-quinoline_MRM.csv')

x_smiles = df_MRM['smiles time']
y_smiles = df_MRM['smiles counts']
x_4_Cl = df_MRM['4-cl-quin time']
y_4_Cl = df_MRM['4-cl-quin counts']
x_7_Aza = df_MRM['7-aza time']
y_7_Aza = df_MRM['7-aza counts']
x_mix = df_MRM['blanks time']
y_mix = df_MRM['blanks counts']

```

```python
plt.figure(figsize = (10,6))
ax1 = plt.subplot()

ax1.plot(x_smiles, y_smiles, color = 'k', label = 'Smiles Analytes')
ax1.set_xlim(200, 400)
plt.legend(frameon = False)
ax1.spines[['top', 'right']].set_visible(False)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Abundance')

plt.savefig('Smiles Analytes MRM.jpeg', dpi = 300)
```

```python
plt.figure(figsize = (10,6))

ax2 = plt.subplot()
ax2.plot(x_4_Cl, y_4_Cl, color = '#ff6361', label = '4-Cl-Quinoline')
# ax2.plot(x_7_Aza, y_7_Aza, color = '#58508d', label = '7-Azaindole')
plt.legend(frameon = False)
ax2.spines[['top', 'right']].set_visible(False)
ax2.set_xlim(100, 300)
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Abundance')

plt.savefig('4-Cl-Quin MRM.jpeg', dpi = 300)
```

```python
plt.figure(figsize = (10,6))
ax3 = plt.subplot()
ax3.plot(x_mix, y_mix, color = 'k', label = '4-Cl-Quinoline')
plt.legend(frameon = False, loc = 'upper right')
ax3.spines[['top', 'right']].set_visible(False)
ax3.set_xlim(15, 210)
ax3.set_ylim(0, 225000)
ax3.set_xlabel('Time (s)')
ax3.set_ylabel('Abundance')

plt.savefig('5 Drops 5 Blanks.jpeg', dpi = 300)
```
