```python

'''Create a function which gets the x and y values from a dataframe. 
This dataframe needs to have chemical name, x coordinate of the well, and y coordinate of the well
to be the columns. Additionally, it would be beneficial to have a column which contains the well name
(ex. well 1)'''

def get_xy_from_other_data(input_chemical): # This outputs as tuples (x,y)
    data_xy = pd.read_csv('xy.csv')
    
    for index, row in data_xy.iterrows():
        if row['chemical name'] == input_chemical:
            return row['x'], row['y']

data_reactions = pd.read_csv('file1.csv') #What is this supposed to contain?

classes = ['Photocatalysts', 'Bases', 'Reagents']

'''iterate through rows in data_reactions dataframe and creates a dictionary of name, x coords, y coords
for each chemical. '''
reactions = []
for index, row in data_reactions.iterrows():
    output_data = {'chemical_name': [], 'x': [], 'y':[]}
    for clas in classes:
        x,y = get_xy_from_other_data(row[clas])
        output_data['chemical_name'].append(row[clas])
        output_data['x'].append(row[clas])
        output_data['y'].append(row[clas])
    reactions.append(output_data)
    
df = pd.DataFrame(reactions)
#What ever gets done with the x,y object?
```
