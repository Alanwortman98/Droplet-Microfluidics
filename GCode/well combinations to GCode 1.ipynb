```python
# photocatalysts, bases, and reagents are a list of x,y coordinates containing each well. 
# Each list is a row in the well plate which corresponds to one variable you want to screen.

PFD_oil = [["x0", "y0"]]
photocatalysts = [["x1", "y1"], ["x2", "y1"], ["x3", "y1"]]
bases = [["x1", "y2"], ["x2", "y2"], ["x3", "y2"]]
reagents = [["x1", "y3"], ["x2", "y3"], ["x3", "y3"]]

# This samples the PFD well at the beginning of the experiment to load PFD into the tubing.

for PFD in PFD_oil:
    print ("G01", PFD, "Z0.0", "F3000")
    print ("G01", PFD, "Z-5.5", "F3000")
    print ("G04", "P10")
    print ("G01", PFD, "Z0.0", "F3000\n")

# This is the loop that makes the combinations of the wells which have been specified. 
# It will sample the PFD well at the end of each loop to space the droplet sets.

for PFD in PFD_oil:
    for R in reagents:
        for B in bases:
            for PC in photocatalysts:
                print ("G01", PC,"Z0.0", "F3000")
                print ("G01", PC,"Z-5.5", "F3000")
                print ("G04 P0.5")
                print ("G01", PC,"Z0.0", "F3000\n")
                
                print ("G01", B,"Z0.0", "F3000")
                print ("G01", B,"Z-5.5", "F3000")
                print ("G04 P0.5")
                print ("G01", B,"Z0.0", "F3000\n")

                print ("G01", R,"Z0.0", "F3000")
                print ("G01", R,"Z-5.5", "F3000")
                print ("G04 P0.5")
                print ("G01", R,"Z0.0", "F3000\n")
                 
                print ("G01", PFD,"Z0.0", "F3000")
                print ("G01", PFD,"Z-5.5", "F3000")
                print ("G04 P2")
                print ("G01", PFD,"Z0.0", "F3000\n")
            

            
```
